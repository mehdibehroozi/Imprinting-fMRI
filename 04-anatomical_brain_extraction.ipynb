# **anatomical brain extraction.**
It is requred to cleand brian mask with hand. The code will creat the anat_brain_mask.nii.gz file. It is not clean yeat. You should clean up the masks by hand to remove eyes from the mask.

```python
import os
import glob
data_path = '/mnt/d/Data/Chick/Imprinting/analysis'
anat_dirs = glob.glob('%s/sub*/anat/anat.nii.gz'%(data_path))
 
for anat_file in anat_dirs:
    print(anat_file)
    
    # bet call edit to use the flags you found worked well on your data
    os.system("bet %s %s/anat_brain -R -f 0.3 -g 0 -m"%(anat_file,anat_file[:-12]))
```

```python
from nilearn.plotting import plot_anat
%matplotlib inline
import matplotlib.pyplot as plt
import os
import glob
data_path = '/mnt/d/Data/Chick/Imprinting/analysis'
subdirs = glob.glob('%s/sub*/anat/anat_brain.nii.gz'%(data_path))
for dir in subdirs:
    plot_anat(dir, title=dir, cut_coords = (20,20,10),
             display_mode='ortho', dim=-1, draw_cross=False, annotate=False)
```
